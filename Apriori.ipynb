```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv('./my_movies.csv')
data.head()
```

```python
mv = data.drop({'V1',"V2", "V3", 'V4','V5'}, axis='columns')
mv
```

```python
mv_a = apriori(mv, min_support=0.05, use_colnames=True)
mv_a
```

```python
mv_ar = association_rules(mv_a, min_threshold=0.6, metric='confidence')
mv_ar
```

```python
mv_finale = mv_ar[(mv_ar.support>0.15)&(mv_ar.confidence>0.5)].sort_values('confidence', ascending=False)
mv_finale
```
